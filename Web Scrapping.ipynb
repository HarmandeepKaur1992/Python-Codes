```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from io import BytesIO
from urllib.parse import urljoin
from PIL import Image
import requests
import certifi
```

```python
from bs4 import BeautifulSoup
```

```python
url="https://www.ebay.com/"
response=requests.get(url, verify=False)

if response.status_code == 200:
    print("Request successful")
else:
    print("Request failed")
    
soup=BeautifulSoup(response.content,'lxml')

#Extracting the heading
headings=soup.find_all('h1')
for h in headings:
    print(h.text)
    
# Extract all paragraphs
paragraphs = soup.find_all('p')
for p in paragraphs:
    print(p.text)

#Extract all image urls
images = soup.find_all('img')
for img in images:
    img_url = img.get('src')  # Use .get to avoid KeyError
    if img_url:
        # Join the base URL with the relative URL to form the complete URL
        img_url = urljoin(url, img_url)
        print(img_url)

        # Fetch the image
        img_response = requests.get(img_url, verify=False)  # Disable SSL verification
       
        img_response.raise_for_status()  # Ensure we got a valid response
        
        # Check if the content type is an image
        content_type = img_response.headers.get('Content-Type')
        if content_type and 'image' in content_type:
            try:
                # Read and display the image
                image = Image.open(BytesIO(img_response.content))
                plt.imshow(image)
                plt.axis('off')  # Hide axes
                plt.show()
            except Exception as e:
                print(f"Could not open image: {e}")
        else:
            print(f"URL does not point to an image: {img_url}")
```
